<h3>Reading the document and abstraction of gene information

```python
with open ('CYCS_dog.txt','r') as file_1:
    string = file_1.read()
    
    #forming a list out of the file given sequence code
    list_strg_1 = string.split('<', 6)
    list_strg_1.pop(0)
    
    #creating tuples inside list with each part(eg. exon,intron) as a element of the list
    exon_intron_tuple = []
    for i in enumerate(list_strg_1):
        exon_intron_tuple.append(i)
   
    #removal of the unwanted CDS gene segment
    exon_intron_tuple.pop(0)
    
    #creating exon and intron list to store respective values seperately
    exon_list = []
    intron_list = []
    
    
    #counting number of exons and introns in the sequence and storing them in respective list seperately
    exon_num = 0
    intron_num = 0
    for item in exon_intron_tuple:
        if 'exon' in item[1]:
            exon_num += 1    
            ex = {}

            #modifying the string in the exon for removal of non-nucleotides
            exon_region = [char for char in item[1] if char == 'A' or char == 'T' or char == 'G' or char == 'C']
            new_exon = exon_region[8:] #as first 8 characters are gene description

            #counting the number of nucleotide
            a_exon= 0
            t_exon = 0
            g_exon = 0
            c_exon = 0

            for nt in new_exon:
                if nt == 'A':
                    a_exon += 1
                elif nt == 'T':
                    t_exon += 1
                elif nt == 'G':
                    g_exon += 1
                elif nt == 'C':
                    c_exon += 1

            #summation of all nucleotide lenghts for finding the lenght of exon
            total_exon = a_exon + t_exon + g_exon + c_exon
            
            #C-G ratio of the exon
            c_g_exon = ((c_exon + g_exon) / total_exon) * 100

            ex = {'count':exon_num, 'A':a_exon, 'C':c_exon, 'T':t_exon, 'G':g_exon, 'Total':total_exon,
                 'C-G-ratio':c_g_exon}

            exon_list.append(ex)

        else:
            intron_num += 1  
            intr = {}
            
            intron_region = [char for char in item[1] if char == 'A' or char == 'T' or char == 'G' or char == 'C']
            new_intron = intron_region[6:]
            
            a_intron = 0
            t_intron = 0
            g_intron = 0
            c_intron = 0 
            
            for nt in new_intron:
                if nt == 'A':
                    a_intron += 1
                elif nt == 'T':
                    t_intron += 1
                elif nt == 'G':
                    g_intron += 1
                elif nt == 'C':
                    c_intron += 1
                    
            total_intron = a_intron + t_intron + g_intron + c_intron
            
            c_g_intron = ((c_intron + g_intron) / total_intron) * 100
            
            intr = {'count':intron_num, 'A':a_intron, 'C':c_intron, 'T':t_intron, 'G':g_intron, 'Total':total_intron,
                   'C-G-ratio':c_g_intron}
            
            intron_list.append(intr)

        
    print(exon_list)
    print(intron_list)
```

<h3>Other parameter required for summary

```python
#total exon lenght
exon_lenght = 0
a_exon = 0
c_exon = 0
t_exon = 0
g_exon = 0
for i in exon_list:
    exon_lenght +=  i['Total']
    a_exon += i['A']
    c_exon += i['C']
    t_exon += i['T']
    g_exon += i['G']

#total intron lenght
intron_lenght = 0
a_intron = 0
c_intron = 0
t_intron = 0
g_intron = 0
for j in intron_list:
    intron_lenght += j['Total']
    a_intron += j['A']
    c_intron += j['C']
    t_intron += j['T']
    g_intron += j['G']

#total gene lenght
gene_lenght = exon_lenght + intron_lenght

#sum of all nucleotides individually in the gene
a = a_exon + a_intron
t = t_exon + t_intron
g = g_exon + g_intron
c = c_exon + c_intron
```

<h3>Writing to a file

```python
with open('Summary.txt', 'w') as result2:
    result2.write("\nCYCS Dog gene description:")
    result2.write("\nThe gene lenght is : " + str(gene_lenght) + (" base pairs"))
    result2.write("\n\nNuleotide Count in gene\n---------------------------")
    result2.write("\nNo. of A's is :" + str(a))
    result2.write("\nNo.of T's is : " + str(t))
    result2.write("\nNo. of G's is : " + str(g))
    result2.write("\nNo. of C's is : " + str(c)) 
    result2.write("\nThe C-G Ratio is : " + str(((c + g) / (gene_lenght)) * 100) + ("%"))
    
    result2.write("\n\n No. of exons is : " + str(exon_num))
    result2.write("\nLenght of exon region is : " + str(exon_lenght) + (" base pairs"))
    
    for i in range(len(exon_list)):
        result2.write("\n\nExon " + str(i) + "\n----------------")
        result2.write("\nNo. of A's is : " + str(exon_list[i]['A']))
        result2.write("\nNo.of T's is : " + str(exon_list[i]['T']))
        result2.write("\nNo. of G's is : " + str(exon_list[i]['G']))
        result2.write("\nNo. of C's is : " + str(exon_list[i]['C']))
        result2.write("\nLenght of exon is : " + str(exon_list[i]['Total']) + (" base pairs"))
        result2.write("\nThe C-G Ratio is : " + str(round(exon_list[i]['C-G-ratio'],2)) + "%")
                 
    result2.write("\n\n No. of introns is : " +str(intron_num))
    result2.write("\nLenght of intron region is : " + str(intron_lenght) + (" base pairs"))
    
    for j in range(len(intron_list)):
        result2.write("\n\nIntron " + str(j) + "\n----------------")
        result2.write("\nNo. of A's is : " + str(intron_list[j]['A']))
        result2.write("\nNo.of T's is : " + str(intron_list[j]['T']))
        result2.write("\nNo. of G's is : " + str(intron_list[j]['G']))
        result2.write("\nNo. of C's is : " + str(intron_list[j]['C']) )
        result2.write("\nLenght of intron is : " + str(intron_list[j]['Total']) + (" base pairs"))
        result2.write("\nThe C-G Ratio is : " + str(round(intron_list[j]['C-G-ratio'],2)) + "%")
    
```
